### Import Libraries:

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

### Loading and Plotting Original Image:
#### Loading the image file into our program by making use of the imread function which takes the path of the file. 

```python
img = cv2.imread("hem.jpeg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

### Converting Image to GrayScale:
#### To decrease the complexity of the image and make the handling easier we will be converting the image to a grayscale image with the help of the cvtColor function.

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(6,6))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

### Inverting the Image:
#### When a image is inverted it helps in studying the image with more precision and detail.

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(6,6))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

#### Smoothen inverted image

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,6))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

### Converting your images to pencil sketches:
#### We will be passing the previous outputs to the function and pass the right and accurate parameters to make the appropriate changes to the image.

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(6,6))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
